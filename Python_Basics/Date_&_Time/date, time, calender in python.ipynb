```python
# Write a Python program to print next 5 days starting from today.
y=(datetime.datetime.now())
x=(datetime.datetime.now()-timedelta(days=1))
z=(datetime.datetime.now()+timedelta(days=1))
print("yesterday was ",x.strftime("%D"))
print("today is ",y.strftime("%D"))
print("tomorrow is ",z.strftime("%D"))
```
