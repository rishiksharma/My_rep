```python
import matplotlib.pyplot as plt
```

```python
values= [1,7,2,8,12,8,3,9,12]
plt.plot(values)
```

```python
v= [1,7,2,8,12,8,3,9,12]
v1=[1,3,5,19,7,6,5,4,12]
plt.plot(v)
plt.plot(v1)
# plt.savefig('Desktop/linechart.jpg')
plt.show()
```

```python
v= [1,7,2,8,12,8,3,9,12]
v1=[1,3,5,19,7,6,5,4,12]
plt.title('multi line comparison', loc='right')
plt.plot(v)
plt.plot(v1)
plt.show()
```

```python
v= [1,7,2,8,12,8,3,9,12]
v1=[1,3,5,19,7,6,5,4,12]
plt.title('multi line comparison', loc='right')
plt.plot(v)
plt.plot(v1)
plt.grid()
plt.show()
```

```python
v= [1,7,2,8,12,8,3,9,12]
v1=[1,3,5,19,7,6,5,4,12]
plt.title('multi line comparison', loc='right')
plt.plot(v)
plt.plot(v1)
plt.xlabel('range')
plt.ylabel('values')
plt.grid()
plt.legend(['first','sec'], loc=2)
plt.show()
```

```python
# annotate
plt.title('multi line comparison', loc='right')
plt.plot(v)
plt.plot(v,v1,'p-b')
plt.xlabel('range')
plt.ylabel('values')
plt.annotate(xy=[1,1],s='first person')
plt.annotate(xy=[3,19],s='sec person')
plt.grid()
plt.legend(['first','sec'], loc=2)
plt.show()
```
