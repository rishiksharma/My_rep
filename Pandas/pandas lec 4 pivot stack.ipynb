```python
import pandas as pd
l1 = ["Red","Green","Orange","Pink","Yellow","White","ben","ten","cat"]
s1= pd.Series(l1)
v=["a","e","i","o","u","A","E","I","O","U"]
l2=[]
s1
for i in range(len(l1)):
    for j in range(len(v)):
        if v[j] in l1:
            l2.append(l1[i])
l2
```
