```python
import pandas as pd
df=pd.read_csv("G:\\other\\weather_data1.csv",parse_dates=['day']) # date conversion from str to timestamp
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True) # affect the original table
df
```

```python
# fill all nan with one specific value
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temprature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

# forward fill and backward fill

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
 new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or␣ ,→"columns"
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

# interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") # considers missing dates .
                                       # it is a neccesity that index is in date format and parsed into timestamp datatype
new_df
```

# drop na

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
#any’ : If any NA values are present, drop that row or column.
#‘all’ : If all values are NA, drop that row or column.
#thresh : int, optional
```

# replace

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weareplace_data.csv")
df
```

```python
# replacing single value
```

```python
new_df = df.replace(-99999, value=np.NaN)
new_df
```

```python
# Replacing list with single value
```

```python
new_df = df.replace(to_replace=[-99999,-88888], value=0)
new_df
```

```python
# Replacing per column (where some value is ok and some other is not)
```

```python
new_df = df.replace({
'temperature': -99999,
'windspeed': -99999,
'event': '0'
}, np.NaN)
new_df
```

```python
# Replacing by using mapping
```

```python
new_df = df.replace({-99999: np.nan,'0': 'Sunny'})
new_df
```

# Regex


```python
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'',regex=True)
new_df
```
